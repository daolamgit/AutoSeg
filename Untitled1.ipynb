```python
##import things
import glob
import os
import h5py
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
test_path = 'Data/Hdf5-channels-last/test'
mask_paths = sorted( glob.glob(os.path.join( test_path, '*.mask')))
img_paths = sorted( glob.glob( os.path.join(test_path, '*.hdf5')))
print mask_paths, img_paths
```

```python
N = 10
n_slice_to_show = 7
n_cols = 3
for i in range( N, N+1):
    h5_mask = h5py.File( mask_paths[i], 'r')
    h5_img = h5py.File( img_paths[i], 'r')
    
    masks_pre = h5_mask['Masks_predict']
    imgs = h5_img['Volume_resize']
    masks_gt = h5_img['Masks_resize']
    
    mid_slice = int( len( masks_gt) / 2)
    plt.subplots( n_slice_to_show * 2, n_cols, figsize = (8 * n_cols, 8 * 2 * n_slice_to_show))
    sub_ind = -1
    for slice in range( mid_slice - n_slice_to_show, mid_slice + n_slice_to_show):
        sub_ind += 1
        mask = np.argmax( masks_pre[slice], axis = -1)
        
#         print masks_pre[slice]
#         print masks_pre[slice].shape
        
        plt.subplot( n_slice_to_show * 2, 3, sub_ind * 3 +1)
        plt.imshow( imgs[slice])
        
        plt.subplot( n_slice_to_show * 2, 3, sub_ind * 3 +2)
        plt.imshow( masks_gt[slice])
        
        plt.subplot( n_slice_to_show * 2, 3, sub_ind * 3 +3)
        plt.imshow( mask)
```
